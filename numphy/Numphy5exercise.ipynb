**A.**

```python
# I Import the NumPy library and give it the alias 'np'
import numpy as np

# I Created a nested Python list (values between 1 and 100)
nested_list = [
    [5, 12, 25],   # First row of values
    [33, 47, 58],  # Second row of values
    [62, 79, 91]   # Third row of values
]

# Convert the nested list into a NumPy ndarray
A = np.array(nested_list)


# using ndarray property (shape)
# check the size/shape of the array (rows and columns)
print(A.shape)


# using ndarray method (max)
# get the highest value inside the array using a method
print("MAXIMUN VALUE USING A.max()",A.max())


# using numpy function (np.max)
# get the highest value again but using numpy function
print("MAXIMUM VALUE USING np.max(A):",np.max(A))
```

**B.**

```python
# create a 5x5 array filled with ones (integer type)
B = np.ones((5, 5), dtype=int)

# check the shape of the array (rows and columns)
print("shape:",B.shape)

# check the total number of elements inside the array
print("size:",B.size)

# check how many dimensions the array has
print("ndim:",B.ndim)
```

**C.**

```python
# create an array with values from -5 to 5

# note: 6 is used because the stop value is not included
C = np.arange(-5, 6)

# multiply each element in the array by 2
C = C * 2

# print the updated array
print("Resulting array:",C)   # expected output: [-10  -8  -6  -4  -2   0   2   4   6   8  10]
```
