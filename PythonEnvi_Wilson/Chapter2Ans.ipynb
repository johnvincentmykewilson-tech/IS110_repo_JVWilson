Write a program that uses input to prompt a user for a number and then outputs the square of the number

Enter a number: 3

The square of 3 is 9

````python
num_str = input("Enter a number:")
# The original code had a SyntaxError because the string was improperly formatted.
# Specifically, 'int' was placed directly inside a double-quoted string without
# proper concatenation or f-string usage.
# Also, 'int' is a built-in Python type; it's better to use a different variable name.
# The input string needs to be converted to an integer before performing multiplication.
num_val = int(num_str)
print(f"The square of {num_val} is {num_val * num_val}")
````

## CHALLENGE

Write a program which prompts the user for a Celsius temperature, convert the temperature to Fahrenheit, and print out the converted temperature.

````python
number = input("Enter a Celsius temperature: ")
fahrenheit = (float(number) * 9/5) + 32
print(f"{number} degrees Celsius is {fahrenheit} degrees Fahrenheit")
````
